```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
X
```

```python
y
```

```python
#test train split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
#imputation
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
model_scores={}
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
#accuracy score for logistic regression
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
a1=accuracy_score(y_test, y_pred)
a1
```

```python

from sklearn.metrics import classification_report


report1 = classification_report(y_test, y_pred)
print(report1)  # Display the classification report

report_dict = classification_report(y_test, y_pred, output_dict=True)  # Convert the report to a dictionary

df = pd.DataFrame(report_dict)  # Create a DataFrame from the report dictionary
df.iloc[:3, :3].T.plot(kind='bar')
plt.show()
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
#Predicting the Test set results
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
#accuracy score for decision tree Classifier
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
a2=accuracy_score(y_test, y_pred)
a2
```

```python

report2 = classification_report(y_test, y_pred)
print(report2)  # Display the classification report

report_dict1 = classification_report(y_test, y_pred, output_dict=True)  # Convert the report to a dictionary

df = pd.DataFrame(report_dict1)  # Create a DataFrame from the report dictionary
df.iloc[:3, :3].T.plot(kind='bar')
plt.show()
```

```python
#random forest
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 7, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
#accuracy score for random forest Classifier
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
#Accuracy score for Random forest
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
a3=accuracy_score(y_test, y_pred)
a3
```

```python

report3 = classification_report(y_test, y_pred)
print(report3)  # Display the classification report

report_dict2 = classification_report(y_test, y_pred, output_dict=True)  # Convert the report to a dictionary

df = pd.DataFrame(report_dict2)  # Create a DataFrame from the report dictionary
df.iloc[:3, :3].T.plot(kind='bar')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=2)

# Train the classifier
knn.fit(X_train, y_train)

# Predict on the test set
y_pred = knn.predict(X_test)

# Calculate and print the accuracy score
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python

report5 = classification_report(y_test, y_pred)
print(report5)  # Display the classification report

report_dict4 = classification_report(y_test, y_pred, output_dict=True)  # Convert the report to a dictionary

df = pd.DataFrame(report_dict2)  # Create a DataFrame from the report dictionary
df.iloc[:3, :3].T.plot(kind='bar')
plt.show()
```

```python
from sklearn.model_selection import  GridSearchCV
# Hyperparameter Tuning
param_grid = {'n_neighbors': [3, 2, 7]}  # Try different values of k
knn = KNeighborsClassifier()
grid_search = GridSearchCV(knn, param_grid, cv=5)
grid_search.fit(X_train, y_train)
best_k = grid_search.best_params_['n_neighbors']

knn = KNeighborsClassifier(n_neighbors=best_k)
knn.fit(X_train, y_train)


y_pred = knn.predict(X_test)

# Calculate and print the accuracy score
a4 = accuracy_score(y_test, y_pred)
a4
```

```python

report4 = classification_report(y_test, y_pred)
print(report4)  # Display the classification report

report_dict3 = classification_report(y_test, y_pred, output_dict=True)  # Convert the report to a dictionary

df = pd.DataFrame(report_dict2)  # Create a DataFrame from the report dictionary
df.iloc[:3, :3].T.plot(kind='bar')
plt.show()
```

```python
accuracy_df = pd.DataFrame({
    'Model': ['logistic', 'Decision ', 'RandomForest','KNN'],
    'Accuracy': [a1, a2, a3, a4]
})
```

```python
accuracy_df.plot(x='Model', y='Accuracy', kind='bar', legend=False)
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Accuracy Comparison for Three Models')
plt.ylim(0, 1)  # Set the y-axis limits between 0 and 1
plt.show()
```

```python
k_values = range(1, 21)  # Try k from 1 to 20
accuracy_scores = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)

# Plotting the accuracy scores for different values of k
plt.plot(k_values, accuracy_scores, marker='o')
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.title('Accuracy vs. k in KNN')
plt.show()
```
